This notebook demonstrates a simple linear regression analysis using R to model Salary based on Years of Experience.

```r
dataset <- read.csv("../manual/regression_data.csv")
```

```r
dataset
```

```r
plot(dataset$YearsExperience, dataset$Salary, col="red",
     main="Salary vs YearsExperience",
     xlab="YearsExperience", ylab="Salary")
```

```r
model <- lm(Salary ~ YearsExperience, data=dataset)
```

```r
library(ggplot2)
ggplot(dataset, aes(x = YearsExperience, y = Salary)) +
  geom_point(color = "red") +
  geom_smooth(method = "lm", color = "blue", se = FALSE) +
  ggtitle("Salary vs YearsExperience") +
  xlab("YearsExperience") +
  ylab("Salary")
```

```r
summary(model)
```
