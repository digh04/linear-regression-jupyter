This notebook demonstrates a simple linear regression analysis using Python to model **Salary** based on **YearsExperience**.

## What is being calculated

This notebook fits a straight line to the data and reports:

- **Slope** — the change in predicted Salary for each additional year of experience
- **Intercept** — the predicted Salary when YearsExperience is 0
- **Correlation coefficient (Pearson's r)** — strength and direction of the linear association between YearsExperience and Salary (values near 1 indicate a strong positive relationship)
- **Mean Squared Error (MSE)** — average squared difference between observed Salary and predicted Salary: MSE = (1/n) * sum((y - y_hat)^2)

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
from sklearn.metrics import mean_squared_error
```

```python
dataset = pd.read_csv("../manual/regression_data.csv")
dataset
```

```python
x = dataset["YearsExperience"]
y = dataset["Salary"]
```

```python
slope, intercept, r_value, p_value, std_err = linregress(x, y)
y_pred = slope * x + intercept
mse = mean_squared_error(y, y_pred)

print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"Correlation coefficient (Pearson's r): {r_value:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
```

## Interpretation

- **Slope:** A slope near 8285 means salary increases by about $8,285 for each additional year of experience, on average.
- **Intercept:** An intercept near 29204 is the model's predicted salary at zero years of experience. Treat this cautiously because zero years is outside the observed range of the data.
- **Correlation (r):** A value near 0.89 indicates a strong positive linear relationship: employees with more experience tend to earn higher salaries.
- **MSE:** A large MSE (on the order of millions) reflects squared salary units, so bigger prediction errors contribute much more to the total. Lower MSE would indicate predictions closer to the observed salaries.

```python
plt.scatter(x, y, color="red", label="Observed data")
plt.plot(x, y_pred, color="blue", label="Fitted line")
plt.text(
    x.min() + 0.1,
    y.max() * 0.92,
    f"Equation: y = {slope:.2f}x + {intercept:.2f}\n"
    f"Correlation coefficient (r): {r_value:.2f}",
    fontsize=10,
    bbox={"facecolor": "white", "alpha": 0.8, "edgecolor": "gray"},
)
plt.title("Salary vs YearsExperience")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.legend()
plt.tight_layout()
plt.show()
```
