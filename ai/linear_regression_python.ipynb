This notebook demonstrates a simple linear regression analysis using Python to model Salary based on Years of Experience.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv("../manual/regression_data.csv")
```

```python
dataset
```

```python
x = dataset["YearsExperience"]
y = dataset["Salary"]
```

```python
plt.scatter(x, y, color="red")
plt.title("Salary vs YearsExperience")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```

```python
model = LinearRegression()
model.fit(dataset[["YearsExperience"]], dataset[["Salary"]])
```

```python
plt.scatter(x, y, color="red")
plt.plot(x, model.predict(dataset[["YearsExperience"]]), color="blue")
plt.title("Salary vs YearsExperience")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```

```python
model.score(dataset[["YearsExperience"]], dataset[["Salary"]])
```
