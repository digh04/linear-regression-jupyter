This notebook demonstrates a simple linear regression analysis using R to model **Salary** based on **YearsExperience**.

## What is being calculated

This notebook fits a straight line to the data and reports:

- **Slope** â€” the change in predicted Salary for each additional year of experience
- **Intercept** â€” the predicted Salary when YearsExperience is 0
- **Correlation coefficient (Pearson's r)** â€” strength and direction of the linear association between YearsExperience and Salary
- **Mean Squared Error (MSE)** â€” average squared difference between observed Salary and predicted Salary

```r
dataset <- read.csv("../manual/regression_data.csv")
dataset
```

```r
model <- lm(Salary ~ YearsExperience, data = dataset)
slope <- coef(model)[2]
intercept <- coef(model)[1]
r <- cor(dataset$YearsExperience, dataset$Salary)
pred <- predict(model)
mse <- mean((dataset$Salary - pred)^2)

cat(sprintf("Slope: %.2f\n", slope))
cat(sprintf("Intercept: %.2f\n", intercept))
cat(sprintf("Correlation coefficient (Pearson's r): %.2f\n", r))
cat(sprintf("Mean Squared Error (MSE): %.2f\n", mse))
```

## Interpretation

- **Slope:** A slope near 8285 means salary increases by about $8,285 for each additional year of experience, on average.
- **Intercept:** An intercept near 29204 is the model's predicted salary at zero years of experience. Treat this cautiously because zero years is outside the observed range of the data.
- **Correlation (r):** A value near 0.89 indicates a strong positive linear relationship: employees with more experience tend to earn higher salaries.
- **MSE:** Because MSE uses squared salary units, larger prediction errors dominate the metric. Lower MSE would indicate predictions closer to the observed salaries.

```r
library(ggplot2)

ggplot(dataset, aes(x = YearsExperience, y = Salary)) +
  geom_point(color = "red") +
  geom_smooth(method = "lm", color = "blue", se = FALSE) +
  annotate(
    "text",
    x = min(dataset$YearsExperience) + 0.1,
    y = max(dataset$Salary) * 0.92,
    label = paste0(
      "Equation: y = ", round(slope, 2), "x + ", round(intercept, 2),
      "\nCorrelation coefficient (r): ", round(r, 2)
    ),
    hjust = 0,
    size = 3.5
  ) +
  ggtitle("Salary vs YearsExperience") +
  xlab("YearsExperience") +
  ylab("Salary")
```
